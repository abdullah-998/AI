```python
import numpy as np
import matplotlib.pyplot as plt

x= np.arange(-20,21,1)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure(figsize=(20,5))
    plt.plot(x,y,'o--')
    plt.grid(True)
    plt.xlabel('X')
    plt.ylabel(ylabel)
    plt.show()
```

```python
def step(x):
    return 0 if x<0.5 else 1

y=[]
for i in x:
    y.append(step(i))
    
plot_graph(y,"STEP")
```

```python
y=1/(1+np.exp(-x))

plot_graph(y,"SIGMOID")
```

```python
y=(np.exp(2*x)-1)/(np.exp(2*x)+1)
plot_graph(y,'Tanh')
```

```python
plot_graph(list(map(lambda a:a if a>=0 else 0,x)),'ReLU')
```

```python
plot_graph(list(map(lambda a:a if a>=0 else a*0.1,x)),'LReLU')
```
