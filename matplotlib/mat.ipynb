```python
W = [random.random(), random.random()]
b = random.random()

print(W, b)

lr=0.05
theta=0.5
```

```python
epoch = 0
count = 0

while count < 4:
    total_error = 0
    print(f"Epoch: {epoch}")
    for i in range(len(x1)):
        weighted_sum = W[0]*x1[i] + W[1]*x2[i] + b
        pred = 0
        if weighted_sum > theta:
            pred = 1
        
        error =  y[i] - pred
        if error != 0:
            count = 0
            total_error += math.sqrt(error**2)
            
            W[0] += lr * error * x1[i]
            W[1] += lr * error * x2[i]
            b    += lr * error
        else:
            count += 1

        print(f"[{x1[i]} , {x2[i]}] = {pred}")
    
    print(f"Loss: {total_error}")
    print("===------------------------------===")
    epoch += 1
```
